```python
#import the necessary libraries

import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
# loading dataset

data = pd.read_csv('language.csv')
data
```

Converting text to array

```python
# converting text and language to array of words
x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
print(x)
```

```python
print(y)
```

```python
data
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x)  # Fit and transform the text data
```

```python
#Splitting the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train
```

```python
print(X_train)
```

```python
y_test
```

Model Training 

```python
# Model training using Multinomial Naive Bayes
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
# Model accuracy on test data

model.score(X_test, y_test)
```

```python
# Accuracy on the testing data
print("Model Accuracy:", model.score(X_test, y_test))
```

```python
#User input for language detection

user_input = input("Enter text to detect language: ")
data = cv.transform([user_input]).toarray()
prediction = model.predict(data)
print("Predicted Language:", prediction[0])
```
